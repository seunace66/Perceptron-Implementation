
### Single-layer Perceptron

The idea is that the algorithm is able to find a set of ideal weights, so that the network correctly classifies the inputs.

For this, if the network response is not the expected response, a weight adjustment is performed.

Network features:

• Linear classifier, ie, performs classification of linearly separable sets (labeled in this example as -1 and 1).
• Supervised learning.

```python
# Weights (Synapses)
w = [0,0]  #Inittialise the randomweight with random value

# Theshold
threshold = 0 # this is for activation whatever I get I want to apply some threshold to squash the output

# Bias
bias = 1

# Learning rate - [0, 1]
learning_rate = 1  # this is just a demo as per real situation it has to be as low as possible

# Maximum number allowed of iterations
max_iterations = 100  #100 Iterations
```

#**Helper-function for plotting red and blue points**

Stores the x and y coordinates for each point of the graph in their respective vectors.

'''below is a user defined(function get_points_of _color, it collect 2 parameter, data is the input variable and label is the target variable, praparing the x and y coordinate value. spliting the input values with the comma delimeter(,)) first position[0] goes to the x, and 2nd position[1] goes to y cordinate'''

```python
def get_points_of_color(data, label):
    x_coords = [float(point.split(",")[0]) for point in data.keys() if data[point] == label]
    y_coords = [float(point.split(",")[1]) for point in data.keys() if data[point] == label]
    return x_coords, y_coords
```

```python
below representation is for manually creating the iteration, how the perceptron process happens, show the perceptron algorithm works(in the keras coding or real coding of Tensorflow 2.2 we wont do this maually, there is a ready made built in API to take care of this)
```

```python
# Turn on the interactive graphics mode
plt.ion()

for k in range(1, max_iterations): #''Initaited a for loop for 100 itereation'''
    hits = 0
    print("\n------------------------- ITERATION "+str(k)+" ------------------------- ")

    for i in range(0,len(x)): #''Pick up one of the records'''
        sum = 0

        # Weighted sum
        for j in range(0,len(x[i])-1):
            sum += x[i][j] * w[j]#''take each value and multiply with the weight'''

        # Output = Bias + Weighted sum
        output = bias + sum #''sum  above plus bias to get the output. the utput need to be squashed to the activation funtion'''

        # Output is determined by the Threshold
        if output > threshold: #''if tthe output is grester than zero y=1 if < than 0 y=-1(here squashing to the classification'''
            y = 1
        else:
            y = -1     

        # Update the Weights if the output does not match with the Desired output
        if y == x[i][2]:
            hits += 1
            answer = "Correct!"
        else:
            for j in range (0,len(w)):       #calculate weight for each features
                w[j] = w[j] + (learning_rate * x[i][2] * x[i][j])
            bias = bias + learning_rate * x[i][2]
            answer = "Error - Updating weight to: "+str(w)

        # Prints the answer
        if y == 1:
            print("\n"+answer)
        elif y == -1:
            print("\n"+answer)

        # Plot the graph
        plt.clf() # Clear figure
        plt.title('Iteration %s\n' % (str(k)))
        plt.grid(False) # Plot a grid
        plt.xlim(-1,1) # Set x-axis limits
        plt.ylim(-1,1) # Set y-axis limits

        xA = 1
        xB = -1

        if w[1] != 0:
            yA = (- w[0] * xA - bias) / w[1]
            yB = (- w[0] * xB - bias) / w[1]
        else:
            xA = - bias / w[0]
            xB = - bias / w[0]

            yA = 1
            yB = -1

        # Plot the black line, that is, we want to learn the black line as faithfully as possible
        #("Best" Decision Boundary)
        plt.plot([0.77, -0.55], [-1, 1], color='k', linestyle='-', linewidth=1)

        # Generates our green line, that is, our learning line (Decision Boundary)
        plt.plot([xA, xB], [yA, yB], color='g', linestyle='-', linewidth=2)

        # Plot blue points
        x_coords, y_coords = get_points_of_color(data_dictionary, '-1')
        plt.plot(x_coords, y_coords, 'bo')

        # Plot red points
        x_coords, y_coords = get_points_of_color(data_dictionary, '1')
        plt.plot(x_coords, y_coords, 'ro')

        # Highlights the current point
        if answer == 'Correct!':
            # Correct - with green color
            plt.plot(x[i][0], x[i][1], 'go', markersize=15, alpha=.5)
        else:
            # Incorrect - with magenta color
            plt.plot(x[i][0], x[i][1], 'mo', markersize=30, alpha=.5)

        # Show the plot
        plt.show()

        # We were able to control the loop time, so a figure will be updated and displayed
        plt.pause(0.05)

    if hits == len(x):
        print("\n---------------------------------------------------------------")
        print("\nFunctionality learned with "+str(k)+" iterations!")
        break;
    '''
    else:
        print("\n---------------------------------------------------------------")
        print("\nFunctionality not learned!")
        break;
    '''
    
print("\nDone!\n")
```
